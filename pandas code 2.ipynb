```python
left = {
    'key':['k0','k1','k2','k3'],
    'A':[10,20,30,40],
    'B':[1,2,3,4]
}
left = pd.DataFrame(left)
left
```

```python
right = {
    'key':['k0','k1','k2','k3'],
    'C':[100,200,300,400],
    'D':[11,12,13,14]
}
right = pd.DataFrame(right)
right
```

```python
pd.merge(left, right, on='key')
```

```python
left = {
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
}
left = pd.DataFrame(left, index = ['k0','k1','k2'])
left
```

```python
right = {
    'C':['C0','C2','C3'],
    'D':['D0','D2','D3']
}
right = pd.DataFrame(right, index = ['k0','k2','k3'])
right
```

```python
left.join(right, how = 'inner')
```
